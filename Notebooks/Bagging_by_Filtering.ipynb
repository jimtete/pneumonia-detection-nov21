```python
##Importing the libraries
import pandas as pd
import numpy as np
%cd Bagging
```

```python
##Test read a csv and sort it

test = pd.read_csv("1.csv").sort_values(by='file_name')
file_names = test['file_name']
file_names = file_names.to_numpy()
print(test.to_numpy())
```

```python
##Read all csv files and save them into a numpy array
full_sets = np.ones((4,1168))

for i in range(4):
  j=i+1
  temp = pd.read_csv(str(j)+".csv").sort_values(by="file_name")
  full_sets[i] = temp['class_id'].to_numpy()
```

```python
calculator = np.zeros((3,1168))

for i in range(3):
  for j in range(1168):
    x = full_sets[i,j]
    if x==0:
      calculator[0,j] = calculator[0,j] + 1
    elif x==1:
      calculator[1,j] = calculator[1,j] + 1
    elif x==2:
      calculator[2,j] = calculator[2,j] + 1

```

```python
final_class = np.zeros(1168)
for i in range(1168):
  a,b,c = calculator[:,i]
  if a>=c and a>=b:
    final_class[i] = 0
  elif b>=a and b>=c:
    final_class[i] = 1
  elif c>=a and c>=b:
    final_class[i] = 2
```

```python
rows, cols = (1168, 2)
full_pred = []
for i in range(1168):
  col = []
  for j in range(cols):
      col.append(0)
  full_pred.append(col)

for i in range(1168):
  full_pred[i][0] = file_names[i]
  full_pred[i][1] = int(final_class[i])

print(full_pred)

export = pd.DataFrame(full_pred)
export.columns=['file_name','class_id']
export.to_csv("DetERNET-TypeR-50-STi-AE86-Trueno_5.csv",index = False)
```
